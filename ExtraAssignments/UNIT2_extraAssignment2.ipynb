```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Sample dataset (you can replace these with actual sentences or load from file)
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Artificial intelligence is transforming the world.",
    "Google acquired DeepMind in 2014.",
    "The chef cooked a delicious meal for the guests.",
    "She opened the window because it was too hot."
]
```

```python
from spacy import displacy
for i, sentence in enumerate(sentences, 1):
    print(f"\nSentence {i}: {sentence}")
    doc = nlp(sentence)

    print("Token\t\tHead\t\tDep\t\tPOS")
    print("-" * 50)
    for token in doc:
        print(f"{token.text:<10}\t{token.head.text:<10}\t{token.dep_:<10}\t{token.pos_}")

    # Visualize the dependency tree directly in notebook
    displacy.render(doc, style="dep", options={"compact": True})
```
