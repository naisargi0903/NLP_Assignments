```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from string import punctuation

# Download necessary packages
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('omw-1.4')
```

```python
nltk.download('all')
```

```python
def lesk(context_sentence, ambiguous_word):

    context_words = set(word_tokenize(context_sentence.lower()))

    context_words = context_words - set(punctuation)

    best_sense = None
    max_overlap = 0

    for synset in wordnet.synsets(ambiguous_word):
        signature = set(word_tokenize(synset.definition().lower()))

        for example in synset.examples():
            signature.update(word_tokenize(example.lower()))

        signature = signature - set(punctuation)

        overlap = len(context_words.intersection(signature))

        if overlap > max_overlap:
            max_overlap = overlap
            best_sense = synset

    return best_sense

context_sentence = input("Enter the sentence: ")
ambiguous_word = input("Enter the ambiguous word: ")

best_sense = lesk(context_sentence, ambiguous_word)

if best_sense:
    print(f"\nBest Sense: {best_sense.name()}")
    print(f"Definition: {best_sense.definition()}")
    if best_sense.examples():
        print(f"Example(s): {best_sense.examples()}")
    else:
        print("No examples available.")
else:
    print("No matching sense found.")
```
